# Sprawdzanie statystyk dla różnych hiperparametrów

```python
def get_statictis_for_algorithm(algorithm='Local DISSFCM'):
    
    names = []
    precisions = []
    
    for i, output in enumerate(output_list):
        name = output_list_name[i]
        n_clusters = n_clusters_list[i]
        
        statistics = output[algorithm][3]
    
        precision = statistics['Precision']
    
        # Dodanie wartości do list
        names.append(name)
        precisions.append(precision)

    return names, precisions


def get_statistics_for_non_incrmental_algorithm(algorithm='KNN'):

    names = []
    precisions = []
    
    for i, models in enumerate(models_non_incrmental_list):
        name = output_list_name[i]
        n_clusters = n_clusters_list[i]
        
        statistics = models[algorithm]
    
        precision = statistics['Precision - segments']
    
        # Dodanie wartości do list
        names.append(name)
        precisions.append(precision)

    return names, precisions
```

```python
import matplotlib.pyplot as plt

def plot_statistics_for_algorithm(precisions, algorithm='FCM'):
    # Tworzenie wykresu
    plt.figure(figsize=(8, 5))
    
    # Rysowanie linii z punktami w przezroczystym czerwonym kolorze
    plt.plot(n_clusters_list, precisions, marker='o', linestyle='-', color=(1, 0, 0, 0.5), markersize=8)
    
    plt.xlabel("Number of starting clusters")
    plt.ylabel("Precision")
    plt.title(algorithm)
    
    # Opcjonalnie dodawanie etykiet do punktów
    for i, v in enumerate(precisions):
        plt.text(n_clusters_list[i], v + 0.01, f"{v:.3f}", ha='center', va='bottom')
    
    plt.grid()  # Opcjonalnie dodaje siatkę
    plt.show()
```

```python
algorithm='FCM'

names, precisions = get_statistics_for_non_incrmental_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='FCM_library'

names, precisions = get_statistics_for_non_incrmental_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='Local DISSFCM'

names, precisions = get_statictis_for_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='Local IFCM'

names, precisions = get_statictis_for_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

### Konwergencja

```python
accuracy = []
precision = []
recall = []
f1 = []

for ele in diagnosis_chunk_dissfcm.statistics_data:
    accuracy.append(ele['Accuracy'])
    precision.append(ele['Precision'])
    recall.append(ele['Recall'])
    f1.append(ele['F1-Score'])
```

```python
accuracy_diff = [accuracy[i] - accuracy[i - 1] for i in range(1, len(accuracy))]
precision_diff = [precision[i] - precision[i - 1] for i in range(1, len(precision))]
recall_diff = [recall[i] - recall[i - 1] for i in range(1, len(recall))]
f1_diff = [f1[i] - f1[i - 1] for i in range(1, len(f1))]
```

### Wykresy różnic między kolejnymi iteracjami

```python
# Tworzenie subplotów
fig, axes = plt.subplots(1, 4, figsize=(20, 5), sharey=True)

# Wykres 1: Accuracy
axes[0].plot(range(len(accuracy_diff)), accuracy_diff, marker='o', color="blue")
axes[0].axhline(0, color='red', linestyle='--', linewidth=1)
axes[0].set_title("Delta Accuracy")
axes[0].set_xlabel("Chunk")
axes[0].set_ylabel("Zmiana (\u0394)")

# Wykres 2: Precision
axes[1].plot(range(len(accuracy_diff)), precision_diff, marker='x', color="green")
axes[1].axhline(0, color='red', linestyle='--', linewidth=1)
axes[1].set_title("Delta Precision")
axes[1].set_xlabel("Chunk")

# Wykres 3: Recall
axes[2].plot(range(len(accuracy_diff)), recall_diff, marker='s', color="orange")
axes[2].axhline(0, color='red', linestyle='--', linewidth=1)
axes[2].set_title("Delta Recall")
axes[2].set_xlabel("Chunk")

# Wykres 4: F1
axes[3].plot(range(len(accuracy_diff)), f1_diff, marker='^', color="purple")
axes[3].axhline(0, color='red', linestyle='--', linewidth=1)
axes[3].set_title("Delta F1")
axes[3].set_xlabel("Chunk")

# Dopasowanie przestrzeni między wykresami
plt.tight_layout()
plt.show()
```

### Statystyczna analzia konwergencji

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(accuracy)
print("p-value:", result[1])

if result[1] < 0.05:
    print("Szereg jest stacjonarny (konwergencja osiągnięta).")
else:
    print("Szereg nie jest stacjonarny.")
```

```python
accuracy_std = np.std(accuracy)
precision_std = np.std(precision)
recall_std = np.std(recall)
f1_std = np.std(f1)

print("Odchylenie standardowe Accuracy:", accuracy_std)
print("Odchylenie standardowe Precision:", precision_std)
print("Odchylenie standardowe Recall:", recall_std)
print("Odchylenie standardowe F1:", f1_std)
```

```python
## imports
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
import numpy as np
```

```python

filepath_train = './datasets/BasicMotions/BasicMotions_TRAIN.arff'
filepath_test = './datasets/BasicMotions/BasicMotions_TEST.arff'

def load_dataset_to_dataframe(path):
    """Load data that is stored within the path

    Args:
        path: string with path containing data
    Returns:
        A pandas dataframe with only data inside
    """
    dataset = arff.loadarff(path)
    df = pd.DataFrame(dataset[0])
    return df 

df_train = load_dataset_to_dataframe(filepath_train)
df_test = load_dataset_to_dataframe(filepath_test)
```

```python
#print(df.iloc[0,0].shape[0])

def stack_time_series(df):
    """This function creates new dataframe without segmentation and classification

    Args:
        df: segmented dataframe
    Returns:
        A pandas dataframe with shape [dimensions of input data,original count of time series]
    """
    dimensionality = df.iloc[0,0].shape[0] # assumption that data dimensionality is the number of rows of first element in dataframe
    # also the first column is expected to store data (second stores classification)
    
    # creating new data frame 
    stacked = pd.DataFrame(index=np.arange(dimensionality), columns=np.arange(1))

    labels = []
    # filling it with empty numpy arrays
    for i in range(dimensionality):
        stacked.iloc[i, 0] = np.array([])
    # filling the dataframe with data    
    for index, row in df.iterrows():
        i = 0
        labels.append(row[1])
        for r in row[0]:
             stacked.iloc[i,0] = np.concatenate((stacked.iloc[i,0],np.array(r.tolist())))
             i += 1


    

    
    return stacked, np.array(labels)
   
stacked_train, y_train = stack_time_series(df)
stacked_test, y_test = stack_time_series(df)



# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Plot different data on each subplot
axs[0, 0].plot(stacked_train.iloc[0,0])
axs[0, 0].set_title('accelerometer X')

axs[0, 1].plot(stacked_train.iloc[1,0])
axs[0, 1].set_title('accelerometer Y')

axs[0, 2].plot(stacked_train.iloc[2,0])
axs[0, 2].set_title('accelerometer Z')

axs[1, 0].plot(stacked_train.iloc[3,0])
axs[1, 0].set_title('gyroscope X')

axs[1, 1].plot(stacked.iloc[4,0])
axs[1, 1].set_title('gyroscope Y')

axs[1, 2].plot(stacked_train.iloc[5,0])
axs[1, 2].set_title('gyroscope Z')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
def stack_data(df):
    # Zakadajc, e kady wiersz w df['relationalAtt'] ma 6 wymiarw
    num_dimensions = 6
    
    # Tworzymy list do przechowywania danych dla nowego DataFrame
    new_data = [[] for _ in range(num_dimensions)]
    
    # Iterujemy przez wszystkie wiersze w df['relationalAtt']
    for row in df['relationalAtt']:
        # Sprawdzamy, czy dugo row jest zgodna z oczekiwaniami
        if len(row) != num_dimensions:
            raise ValueError(f"Oczekiwana dugo wiersza: {num_dimensions}, uzyskana: {len(row)}")
        
        # Dodajemy kady wymiar do odpowiedniego miejsca w new_data
        for i in range(num_dimensions):
            new_data[i].append(row[i])
    
    # Tworzymy nowy DataFrame
    new_df = pd.DataFrame(new_data).transpose()  # transponujemy, aby kady wiersz mia 6 wymiarw
    new_df.columns = [f'dimension_{i+1}' for i in range(num_dimensions)]
    dfs = {}

    return new_df

# Scalanie wierszy
#result_df = stack_data(df)
print(df['relationalAtt'].iloc[0].shape)
```

```python
# Spaszczenie danych
def FlattenData(result_df):
    flattened_data = []
    for i in range(result_df.shape[0]):
        row = []
        for col in result_df.columns:
            row.extend(result_df.iloc[i][col])
        flattened_data.append(row)

    flattened_data = np.array(flattened_data)

    print(flattened_data.shape)  
    return flattened_data
```

```python
flattened_data_train = FlattenData(result_df_train)
flattened_data_test = FlattenData(result_df_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
flattened_data_train = scaler.fit_transform(flattened_data_train)
```

```python
from libraries.tests.test_dataset import test_dataset
import numpy as np
from libraries.plot_functions import plot_centroids_from_dict, plot_metrics_by_algorithm, plot_metrics_by_metrics, plot_metrics_by_dataset
```

```python
output_list = []
models_non_incrmental_list = []
output_list_name = []
```

### Testowanie hyper parametrów

```python
n_clusters_list = list(range(4, 17, 2))
```

```python
chunk_length_train = 100
chunk_length_test = 100
elements_in_class_train = 0
elements_in_class_test = 0
n_clusters = 8
n_classes = 4
dim = 6
m = 2
g_error = 0.05
injection = 0.0
dataset_name = 'BasicMotions'
visualise_processed_data = False
visualise_non_incremental_data = False
visualise_incremental_data = False
print_statistics_incremental_data = False
visualise_output_of_incremental_data = False
```

```python
for n_clusters in n_clusters_list:
    output, models_non_incrmental = test_dataset(chunk_length_train = chunk_length_train, chunk_length_test = chunk_length_test, elements_in_class_train  = elements_in_class_train, n_clusters = n_clusters, n_classes = n_classes, dim = dim,  m = m, error = g_error, injection = injection, dataset_name = dataset_name, visualise_processed_data = visualise_processed_data, visualise_non_incremental_data = visualise_non_incremental_data, visualise_incremental_data = visualise_incremental_data, visualise_output_of_incremental_data = visualise_output_of_incremental_data,  print_statistics_incremental_data = print_statistics_incremental_data)
    output_list.append(output)
    models_non_incrmental_list.append(models_non_incrmental)
    output_list_name.append(dataset_name + f' n_clusters={n_clusters}')
    plot_metrics_by_dataset([output], [dataset_name])
```

```python
def get_statictis_for_algorithm(algorithm='Local DISSFCM'):
    
    names = []
    precisions = []
    
    for i, output in enumerate(output_list):
        name = output_list_name[i]
        n_clusters = n_clusters_list[i]
        
        statistics = output[algorithm][3]
    
        precision = statistics['Precision']
    
        # Dodanie wartości do list
        names.append(name)
        precisions.append(precision)

    return names, precisions


def get_statistics_for_non_incrmental_algorithm(algorithm='KNN'):

    names = []
    precisions = []
    
    for i, models in enumerate(models_non_incrmental_list):
        name = output_list_name[i]
        n_clusters = n_clusters_list[i]
        
        statistics = models[algorithm]
    
        precision = statistics['Precision - segments']
    
        # Dodanie wartości do list
        names.append(name)
        precisions.append(precision)

    return names, precisions
```

```python
import matplotlib.pyplot as plt

def plot_statistics_for_algorithm(precisions, algorithm='FCM'):
    # Tworzenie wykresu
    plt.figure(figsize=(8, 5))
    
    # Rysowanie linii z punktami w przezroczystym czerwonym kolorze
    plt.plot(n_clusters_list, precisions, marker='o', linestyle='-', color=(1, 0, 0, 0.5), markersize=8)
    
    plt.xlabel("Number of starting clusters")
    plt.ylabel("Precision")
    plt.title(algorithm)
    
    # Opcjonalnie dodawanie etykiet do punktów
    for i, v in enumerate(precisions):
        plt.text(n_clusters_list[i], v + 0.01, f"{v:.3f}", ha='center', va='bottom')
    
    plt.grid()  # Opcjonalnie dodaje siatkę
    plt.show()
```

```python
algorithm='FCM'

names, precisions = get_statistics_for_non_incrmental_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='FCM_library'

names, precisions = get_statistics_for_non_incrmental_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='Local DISSFCM'

names, precisions = get_statictis_for_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```

```python
algorithm='Local IFCM'

names, precisions = get_statictis_for_algorithm(algorithm=algorithm)

plot_statistics_for_algorithm(precisions, algorithm)
```
